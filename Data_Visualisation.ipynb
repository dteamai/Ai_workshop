# Data Visualization with matplotlib



Matplotlib is a Python library specializing in the development of two-dimensional charts

**First Plot**

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1,2,3,4]) # by default x value is generated by matplotlib if x value is not given
plt.show()
```

**matplotlib.pyplot.plot**

**Format Strings**

* A format string consists of a part for color, marker and line:
    * fmt = '[color][marker][line]'

```python
plt.plot([1,2,3,4],[1,4,9,16],'ro')
```

**Axis Range**

we can define the range both on the x axis and on the y axis by defining the details of a list [xmin, xmax, ymin, ymax] and then passing it as an argument to the axis() function.

```python
plt.plot([1,2,3,4],[1,4,9,16],'ro')
plt.axis([0,5,0,20])
plt.title('sample')
```

**Line Width**

* By specifying the linewidth, the thickness of line in the plot can be changed.

```python
plt.plot([1,2,4,2,1,0,1,2,1,4],linewidth=2.0)
```

## Working with Multiple Figures and Axes



The subplot() function, in addition to subdividing the figure in different drawing areas, is used at the same time in order to focus the commands on a specific subplot.

The argument of the subplot() function is composed of three integers. The first number defines how many parts the figure is split into vertically. The second number defines how many parts the figure is divided into horizontally. The third issue selects which is the current subplot on which we can direct commands.

```python
t = np.arange(0,5,0.1)
y1 = np.sin(2*np.pi*t)
y2 = np.sin(2*np.pi*t)
plt.subplot(122)
plt.plot(t,y1,'bo')
plt.subplot(121)
plt.plot(t,y2,'ro')
```

Now we do the same thing by dividing the figure in two vertical subplots. The numbers to be passed as arguments to the subplot() function are 121 and 122

```python
t = np.arange(0.,1.,0.05)
t = np.arange(0,5,0.1)
y1 = np.sin(2*np.pi*t)
y2 = np.sin(2*np.pi*t)
plt.subplot(121)
plt.plot(t,y1,'b-.')
plt.subplot(122)
plt.plot(t,y2,'r--')
```

#### Adding a Grid



```python
plt.axis([0,5,0,20])
plt.title('My first plot',fontsize=20,fontname='Times New Roman')
plt.xlabel('Counting',color='gray')
plt.ylabel('Square values',color='gray')
plt.text(1,1.5,'First')
plt.text(2,4.5,'Second')
plt.text(3,9.5,'Third')
plt.text(4,16.5,'Fourth')
plt.text(1.1,12,r'$y = x^2$',fontsize=20,bbox={'facecolor':'yellow','alpha':0.2})
plt.grid(True)
plt.plot([1,2,3,4],[1,4,9,16],'ro')
```

#### Adding a Legend



```python
plt.axis([0,5,0,20])
plt.title('My first plot',fontsize=20,fontname='Times New Roman')
plt.xlabel('Counting',color='gray')
plt.ylabel('Square values',color='gray')
plt.text(1,1.5,'First')
plt.text(2,4.5,'Second')
plt.text(3,9.5,'Third')
plt.text(4,16.5,'Fourth')
plt.grid(True)
plt.plot([1,2,3,4],[1,4,9,16],color='r')
plt.plot([1,2,3,4],[1,4,9,16],'ro')
plt.plot([1,2,3,4],[0.8,3.5,8,15],color='g')
plt.plot([1,2,3,4],[0.8,3.5,8,15],'g^')
plt.plot([1,2,3,4],[0.5,2.5,4,12],'b*')
plt.plot([1,2,3,4],[0.5,2.5,4,12],color='b')
plt.legend(['First series','Second series','Third series'],loc=2)
```

# Advanced Visualization using seaborn
* seaborn is a library that builds on top of Matplotlib and Pandas to provide easy-to-use statistical plotting routines. In this recipe, we give a few examples, adapted from the official documentation, of the types of statistical plot that can be created with seaborn.



```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

* seaborn comes with built-in datasets, which are useful when making demos. The tips dataset contains bills and tips for taxi journeys:


```python
tips = sns.load_dataset('tips')
tips.head()
```

* seaborn implements easy-to-use functions to visualize the distribution of datasets. Here, we plot the histogram, Kernel Density Estimation (KDE), and a gamma distribution fit for our dataset:

    * In statistics, kernel density estimation (KDE) is a non-parametric way to estimate the probability density function of a random variable. Kernel density estimation is a fundamental data smoothing problem where inferences about the population are made, based on a finite data sample.

    * The gamma distribution is the maximum entropy probability distribution

```python
# We create two subplots sharing the same y axis.
f, (ax1, ax2) = plt.subplots(1, 2,figsize=(12, 5),sharey=True)
    
# Left subplot.
# Histogram and KDE (active by default).
sns.distplot(tips.total_bill,ax=ax1,hist=True)
    
# Right subplot.
# "Rugplot", KDE, and gamma fit.
sns.distplot(tips.total_bill,ax=ax2,hist=False,kde=True,rug=True,fit=stats.gamma,fit_kws=dict(label='gamma'),kde_kws=dict(label='kde'))
ax2.legend()
```

**sns.distplot()**

* We can make a quick linear regression to visualize the correlation between two variables:


```python
sns.regplot(x="total_bill", y="tip", data=tips)
```

**sns.regplot**

* We can also visualize the distribution of categorical data with different types of plot. 
* Here, we display a bar plot, a violin plot, and a swarm plot that show an increasing amount of details:


```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12, 4), sharey=True)
sns.barplot(x='sex', y='tip', data=tips, ax=ax1)
sns.violinplot(x='sex', y='tip', data=tips, ax=ax2)
sns.swarmplot(x='sex', y='tip', data=tips, ax=ax3)
```

* The bar plot shows the mean and standard deviation of the tip, for males and females.
* The violin plot shows an estimation of the distribution in a more informative way than the bar plot, especially with non-Gaussian or multimodal distributions.
* The swarm plot displays all points, using the non-informative x axis to make them non-overlapping.

* FacetGrid lets us explore a multidimensional dataset with several subplots organized within a grid.
* Here, we plot the tip as a function of the bill, with a linear regression, for every combination of smoker (Yes/No) and sex (Male/Female):


```python
g = sns.FacetGrid(tips, col='smoker', row='sex')
g.map(sns.regplot, 'total_bill', 'tip')
```
